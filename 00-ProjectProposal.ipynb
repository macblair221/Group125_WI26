## Background and Prior Work

Emotion has a significant impact on cognitive processes in humans, including attention, learning, and memory. <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) The amygdala is a brain structure located in the temporal lobe, and it is responsible for processing emotions, such as fear, anger, and anxiety. The hippocampal complex is also located in the temporal lobe, and it is responsible for memory encoding. 

A previous study has revealed how the amygdala and the hippocampal complex work together when forming emotional memories. <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) While the two brain structures are linked to two independent memory systems, the amygdala has shown to have influence on the encoding and storage of memories dependent on the hippocampal complex. The ability to attend to a stimulus is a primary factor in how well the stimulus is encoded. Since the amygdala responds to emotional stimuli rapidly, emotional events can receive priority when memories are encoded. Additionally, when memories are consolidated, events that elicit emotional responses are more likely to be remembered for later. 

In a study that experimented with autobiographical memories that varied in valence (positive or negative) and intensity (high or low), it was found that the emotional intensity affects the properties of autobiographical memories more than valence. <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) The study concluded that highly intense events tend to be remembered longer and with greater vividness. 

Another study has also found that a perspective change from a field (first-person) to an observer (third-person) reduces the emotional intensity of recalled events. <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) Given the study above, a field-to-observer perspective change is likely to reduce the vividness of a memory. Additionally, the study indicates that the autobiographical memory system has plasticity, and a perspective change alters the autobiographical records of recalled events. 

1. <a name="cite_note-1"></a> [^](#cite_ref-1) Tyng, Chai M et al. “The Influences of Emotion on Learning and Memory.” Frontiers in psychology vol. 8 1454. 24 Aug. 2017, doi:10.3389/fpsyg.2017.01454
2. <a name="cite_note-2"></a> [^](#cite_ref-2) Phelps, Elizabeth A. “Human Emotion and Memory: Interactions of the Amygdala and Hippocampal Complex.” Current Opinion in Neurobiology, vol. 14, no. 2, 2004, pp. 198–202. Current Opinion in Neurobiology, doi:10.1016/j.conb.2004.03.015.
3. <a name="cite_note-3"></a> [^](#cite_ref-3) Talarico, J.M., LaBar, K.S. & Rubin, D.C. Emotional intensity predicts autobiographical memory experience. Memory & Cognition 32, 1118–1132 (2004). https://doi.org/10.3758/BF03196886. 
4. <a name="cite_note-4"></a> [^](#cite_ref-4) Sekiguchi, Tatsuya, and Shogo Nonaka. “The Long-Term Effect of Perspective Change on the Emotional Intensity of Autobiographical Memories.” Cognition and Emotion, vol. 28, no. 2, 2014, pp. 375–383, doi:10.1080/02699931.2013.825233.
